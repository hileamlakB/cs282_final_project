```python
import pandas as pd
from collections import Counter

# Assuming df is your pandas DataFrame
# Step 1 & 2: Create 'time' column and drop 'start_time' and 'end_time'
df['time'] = '['+df['start_time'] + " - " + df['end_time'] + ")"
df.drop(['start_time', 'end_time'], axis=1, inplace=True)

# # Step 3: Replace NaN values
# df.fillna(0, inplace=True)  # Replace NaN with 0 or another suitable value

# Step 4: Delete the 'drugs' column
df.drop('drugs', axis=1, inplace=True)

```

```python
# Step 5: Modify 'combined_actions'
def process_combined_actions(actions):
    # Remove extra spaces and split by comma
    actions = str(actions)
    print(actions)
    actions_list = [action for action in actions.split(',') if action]
    if not actions_list:  # If the list is empty
        return '4'
    most_common = Counter(actions_list).most_common(1)[0][0]
    return most_common

df['combined_actions'] = df['combined_actions'].apply(process_combined_actions)
```

```python
import pandas as pd

# Assuming df is your pandas DataFrame

# Step 1: Rename 'icustay_id' to 'stay_id'
df.rename(columns={'icustay_id': 'stay_id'}, inplace=True)

# Step 2: Make 'time' the second column
time_col = df.pop('time')
df.insert(1, 'time', time_col)

# Step 3: Replace empty entries in 'combined_actions' with '4'
df['combined_actions'] = df['combined_actions'].replace('', '4')

# Step 4: Create a dictionary to map 'combined_actions' to a list of 'total_cost' values
actions_to_costs = {}
for action in df['combined_actions'].unique():
    actions_to_costs[action] = df.loc[df['combined_actions'] == action, 'total_cost'].tolist()

# actions_to_costs now contains the mapping from actions to costs
```

```python
actions_to_costs2 = {key.strip():[value for value in actions_to_costs[key] if value != float('nan')] for key in actions_to_costs if key.strip() and key.strip() != 'nan'}
actions_to_costs2
```

```python
import numpy as np
actions_to_costs = {
    '2': [144106.24999999997, 38041111.111111104, 0.1766666666666666],
    '3': [0.2758, 15.12, 22.15],
    '0': [4225.327777777777, 0.0111, 0.2771979166666666, 0.20128, 0.473333333333332, 1.18333333333333, 0.2758],
    '1': [442.2]
}

# Calculate the average, ignoring NaN values
average_costs = {action: np.nanmean(costs) for action, costs in actions_to_costs.items()}
average_costs['4'] = 0
```

```python
# Assuming df is your pandas DataFrame

# Replace NaN in 'total_cost' with the average cost for the corresponding 'combined_actions'
def fill_nan_with_action_average(row):
    if pd.isna(row['total_cost']):
        return average_costs.get(row['combined_actions'].strip(), np.nan)
    return row['total_cost']

df['total_cost'] = df.apply(fill_nan_with_action_average, axis=1)
```

```python
df = df.sort_values(by='stay_id')
df.to_csv('mimmic_ards_cut2.csv', index=False)
```
